```python
import numpy as np
import pandas as pd
import scipy.stats
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
```

IMPORT DATA

```python
application_test = pd.read_csv("application_test.csv")
application_train = pd.read_csv("application_train.csv")
```

### APPLICATION TRAIN

DELETE COLUMNS

```python
application_train = application_train.drop(application_train.loc[:, 'ORGANIZATION_TYPE':'AMT_REQ_CREDIT_BUREAU_YEAR'].columns, axis = 1)
```

```python
application_train = application_train.drop(columns = ['OWN_CAR_AGE', 'OCCUPATION_TYPE','WEEKDAY_APPR_PROCESS_START'])
```

```python
application_train = application_train[application_train['CODE_GENDER'].str.contains("XNA")==False]
application_train = application_train[application_train['NAME_INCOME_TYPE'].str.contains("Maternity leave")==False]
```

DROP *NaN*

```python
application_train.isna().sum()
```

```python
application_train = application_train.dropna()
```

```python
application_train.isna().sum()
```

```python
application_train.info()
```

DUMMY VARIABLES

```python
dms = pd.get_dummies(application_train[['NAME_CONTRACT_TYPE']])
data_train = pd.concat([application_train,dms], axis=1 )
data_train = data_train.drop(['NAME_CONTRACT_TYPE'],axis=1)
```

```python
dms = pd.get_dummies(application_train[['CODE_GENDER']])
data_train = pd.concat([data_train,dms], axis=1 )
data_train = data_train.drop(['CODE_GENDER'],axis=1)
```

```python
dms = pd.get_dummies(application_train[['FLAG_OWN_CAR']])
data_train = pd.concat([data_train,dms], axis=1 )
data_train = data_train.drop(['FLAG_OWN_CAR'],axis=1)
```

```python
dms = pd.get_dummies(application_train[['FLAG_OWN_REALTY']])
data_train = pd.concat([data_train,dms], axis=1 )
data_train = data_train.drop(['FLAG_OWN_REALTY'],axis=1)
```

```python
dms = pd.get_dummies(application_train['NAME_TYPE_SUITE'])
data_train = pd.concat([data_train,dms], axis=1 )
data_train = data_train.drop(['NAME_TYPE_SUITE'],axis=1)
```

```python
dms = pd.get_dummies(application_train[['NAME_INCOME_TYPE']])
data_train = pd.concat([data_train,dms], axis=1 )
data_train = data_train.drop(['NAME_INCOME_TYPE'],axis=1)
```

```python
dms = pd.get_dummies(application_train[['NAME_EDUCATION_TYPE']])
data_train = pd.concat([data_train,dms], axis=1 )
data_train = data_train.drop(['NAME_EDUCATION_TYPE'],axis=1)
```

```python
dms = pd.get_dummies(application_train[['NAME_FAMILY_STATUS']])
data_train = pd.concat([data_train,dms], axis=1 )
data_train = data_train.drop(['NAME_FAMILY_STATUS'],axis=1)
```

```python
dms = pd.get_dummies(application_train[['NAME_HOUSING_TYPE']])
data_train = pd.concat([data_train,dms], axis=1 )
data_train = data_train.drop(['NAME_HOUSING_TYPE'],axis=1)
```

```python
data_train
```

```python
data_train.info()
```

SPLITTING DATA

```python
Y_train = data_train[['TARGET']]
Y_train
```

```python
X_train = data_train.drop(columns = ['SK_ID_CURR','TARGET'])
X_train
```

LOGISTIC REGRESSION

```python
import statsmodels.api as sm
```

```python
logreg = LogisticRegression()
logreg.fit(preprocessing.scale(X_train), Y_train)
```
